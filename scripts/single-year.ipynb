```python
import json
import pandas as pd
import matplotlib.pyplot as plt

# Read json data
with open('../data/SP500 2019.json') as f:
    json_data = json.load(f)

# Create data frame from data
data = pd.DataFrame(json_data['historical'])

# Sort the date in descending order
data = data.sort_values('date', ascending=True, ignore_index=True)

# Set the index to date
data = data.set_index(pd.DatetimeIndex(data.date))

# Data to plot
data['close'].plot(grid=True, figsize=(9,5), color='orange', linewidth=2)

# Configure the chart
plt.title('SP 500 Index')
plt.ylabel('Index')
plt.xlabel('Date')
plt.show()
```
